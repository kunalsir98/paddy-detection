```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,6))
plt.pie(df['Soil Types'].value_counts(), labels=df['Soil Types'].value_counts().index,autopct='%1.1f%%')
plt.title('Distribution of soil types')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Seedrate(in Kg)',y='Variety',data=df)
plt.title('Distribution of seedrate for different varieties')
plt.xlabel('Seedreate(in kg)')
plt.ylabel('Variety')
plt.show()
```
